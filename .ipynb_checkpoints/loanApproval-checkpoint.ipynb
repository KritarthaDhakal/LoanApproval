### 3.2 Exploring Values for Categorical Columns

```python
categorical_cols = ['person_gender', 'person_education', 'person_home_ownership', 'loan_intent', 'previous_loan_defaults_on_file']

for col in categorical_cols:
    print(f'\n"{col}" has unique values: {df[col].unique()}\n')
```

### 3.3 Stats for Numerical Columns

```python
numerical_cols = list(col for col in df.columns if col not in categorical_cols)
df[numerical_cols].describe()
```

### 3.4 Outlier Detection

```python
len(df[df['person_age'] > 100])
```

```python
df2 = df[~(df['person_age'] > 100)]
df2.shape
```

From the description table above, we see that the 'person_age' has a max value of 144 which is not possible. So, we analyze the data to find ages above 100 and remove them from the dataset.

```python
len(df2[df2['person_income'] > 1000000])
```

```python
df3 = df2[~(df2['person_income'] > 1000000)]
df3.shape
```

From the description table above, we see that the 'person_income' has a max value of over 7 milion which is not usually an outlier. But, the max value for 'loan_amnt' is 35, 000. It is highly unlikely that a person with 7 million income would need require a loan of 35k (a relatively small amount). So, we analyze the data to find income above 1 million and remove them from the dataset. As our dataset has almost 45000 data points, we can remove such extreme cases if even they are real case secnarios. 

```python
len(df3[df3['person_emp_exp'] > 50])
```

```python
df4 = df3[~(df3['person_emp_exp'] > 50)]
df4.shape
```

The 'person_emp_exp' shows the total employed experience years of a person. From the description table above, we see that the 'person_emp_exp' has a max value of 125 which is not possible. Generally the maximum employee experience of a person it 40 to 50 years. So, we analyze the data to find experience years above 50 and remove them from the dataset.

```python
len(df4[df4['loan_amnt'] < 1000])
```

```python
df5 = df4[~(df4['loan_amnt'] < 1000)]
df5.shape
```

Here, I have made a choice to remove data points where loan amounts are lesser than 1000 as that 1000 is regarded as the base loan amount provided by lenders.

## 4. Data Visualization

### 4.1 Value Count for Target Variable Categories 

```python
df5['loan_status'].value_counts().plot(kind='bar')
plt.title('Loan Rejected vs Approved')
plt.xlabel('Loan Approval Status')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

The above bar chart shows the frequency of loan status: Rejected (0) and Approved (1). We see the imbalance in the dataset as the cases for loan approved is way lesser than the cases for loan rejected.

### 4.2 Relationship between Target Variable and Categorical Columns

```python
for col in categorical_cols:
    sns.countplot(x=col, hue='loan_status', data=df5)
    plt.title(f'Loan Status by {col.title()}')
    plt.xlabel(f'{col.title()}')
    plt.ylabel('Count')
    plt.show()
```

* Graph of Loan Status by Gender - we notice an almost even balance for loan approval.
* Graph of Loan Status by Education - we see students in 'High School', 'Asscociate', and 'Bachelors' requiring more loans (students loans probably). Also, a person with a 'Doctorate' is more likely to get his/her loan approved.
* Graph of Loan Status by Home Ownership - we notice poeple with 'Mortgage' and 'Rent' require more loans. Among the two, people with 'Rent' are more likely to get their loans approved.
* Graph of Loan Intent - we notice most loans are approved for 'Medical' and 'Debt Consolidation' and the least for 'Venture'.
* Graph of Previous Loans - people with previous loans will not get another loan approved. 

### 4.3 Relationship between Target Variable and Numerical Columns

```python
for col in ['loan_amnt', 'loan_int_rate', 'loan_percent_income', 'cb_person_cred_hist_length', 'credit_score']:
    mean_scores = df5.groupby('loan_status')[col].mean().reset_index()
    sns.barplot(x='loan_status', y=col, data=mean_scores)
    plt.title(f'Mean {col.title()} by Loan Status')
    plt.xlabel('Loan Status')
    plt.ylabel(f'Mean {col.title()}')
    plt.show()
```

* Graph of Loan Status - People requiring higher loan amounts are generally approved.
* Graph of Loan Interest Rate - Higher Interest Rate loans are approved more.
* Graph of Loan amount as a percentage of annual income - people who received loan approvals are taking on larger loans relative to their income.
* Graph of Credit History - Does not really affect the approval rate.
* Graoh of Credit Score - Does not really affect the approval rate.

### 4.4 Relationship amoung all Features

```python
correlation = df5[numerical_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.xticks(rotation=45)
plt.show()
```

From the correlation matrix, we observe that only Interest Rate and Loan amount as a percentage of annual income has slight correlation with 'Loan Status'.
